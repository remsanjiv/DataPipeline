```python
## Make copies of train and test data ::

train_copy = copy.deepcopy(train_file)
test_copy = copy.deepcopy(test_file)

```

```python
p1 = pdPipeline([('var_select', VarSelector(['Amount.Requested', 'Amount.Funded.By.Investors', 'Open.CREDIT.Lines', 'Revolving.CREDIT.Balance'])), ('convert_to_numeric', convert_to_numeric()),
('missing_str', DataFrameImputer())])
```

```python
p2 = pdPipeline([('var_select',VarSelector(['Debt.To.Income.Ratio'])), ('string_clean', string_clean(replace_it='%', replace_with='')), ('convert_to_numeric', convert_to_numeric()), ('missing_trt', DataFrameImputer())])
```

```python
p3 = pdPipeline([('var_select', VarSelector(['Loan.Length', 'Loan.Purpose', 'State', 'Home.Ownership', 'Employment.Length'])), ('missing_str', DataFrameImputer()), ('create_dummies', get_dummies_Pipe(20))])
```

```python
p4= pdPipeline([('var_select', VarSelector(['Monthly.Income', 'Inquiries.in.the.Last.6.Months'])), ('missing_str', DataFrameImputer())])
```

```python
p5= pdPipeline([('var_select', VarSelector(['FICO.Range'])), ('custom_fico', custom_fico())])
```

```python
data_pipe = FeatureUnion([('obj_to_num', p1),('dtir', p2), ('obj_to_dum', p3), ('num', p4), ('fico', p5)])
```

```python
data_pipe.fit(train_copy)
```
